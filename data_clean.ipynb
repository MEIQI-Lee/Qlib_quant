```python
import csv

# 打开txt文件和csv文件
with open(r'C:\Users\dell\.qlib\qlib_data\kcb_data\instruments\all.txt', 'r', encoding='utf-8') as txt_file, \
open('data.csv', 'w', newline='', encoding='utf-8') as csv_file:
# 创建csv写入器
csv_writer = csv.writer(csv_file)

# 逐行读取txt文件
for line in txt_file:
# 使用空格分割每行的数据
row_data = line.split()
# 将分割后的数据写入csv文件
csv_writer.writerow(row_data)
```

```python
import pandas as pd
# 读取文本文件
with open(r'C:\Users\dell\.qlib\qlib_data\kcb_data\instruments\all.txt', 'r') as file:
    lines = file.readlines()
# 处理数据并分列
data = [line.strip().split() for line in lines]

# 创建 DataFrame
df = pd.DataFrame(data, columns=['stock_code', 'start_time', 'end_time'])

# 保存为 CSV 文件
df.to_csv('all_stock.csv', index=False)
```
